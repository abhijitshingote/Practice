```python
from keras.datasets import mnist

from keras.utils import np_utils
```

```python
(X_train,y_train),(X_test,y_test)=mnist.load_data()
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten

kn=Sequential()
```

```python
kn.add(Flatten(input_shape=X_train.shape[1:]))
kn.add(Dense(256,activation='relu'))
kn.add(Dropout(0.2))
kn.add(Dense(256,activation='relu'))
kn.add(Dropout(0.2))
kn.add(Dense(10,activation='softmax'))
```

```python
kn.summary()
```

```python
new_y_train=np_utils.to_categorical(y_train)
new_X_train=X_train/255
new_y_test=np_utils.to_categorical(y_test)
new_X_test=X_test/255
```

```python
kn.compile(loss='categorical_crossentropy', optimizer='rmsprop', 
              metrics=['accuracy'])
```

```python
from keras.callbacks import ModelCheckpoint

checkpointer = ModelCheckpoint(filepath='mnist.model.best.hdf5', 
                               verbose=1, save_best_only=True)

kn.fit(new_X_train,new_y_train,batch_size=128,epochs=10,verbose=1,validation_split=0.2,
       shuffle=True,callbacks=[checkpointer])
```

```python
# load the weights that yielded the best validation accuracy
kn.load_weights('mnist.model.best.hdf5')
# evaluate test accuracy
score = kn.evaluate(new_X_test, new_y_test, verbose=0)
accuracy = 100*score[1]

# print test accuracy
print('Test accuracy: %.4f%%' % accuracy)
```
